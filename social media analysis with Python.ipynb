```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
```

```python
data = {'Date': pd.date_range('2021-01-01', periods=500),'Category': [random.choice(categories) for _ in range(500)],'Likes': np.random.randint(0, 10000, size=500)}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Date'], df['Likes'])
plt.title('Scatter Plot of Likes Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Likes')
plt.show()
```

```python
plt.figure(figsize=(8,8))
df['Category'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Category Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Category', y='Likes', estimator=sum, data=df)
plt.title('Total Likes per Category')
plt.xlabel('Category')
plt.ylabel('Total Likes')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df.groupby('Date')['Likes'].sum().plot(kind='line')
plt.title('Total Likes Over Time')
plt.xlabel('Date')
plt.ylabel('Total Likes')
plt.grid(True)
plt.show()
```
